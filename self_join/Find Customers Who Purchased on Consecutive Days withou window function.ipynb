```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, date_add
from pyspark.sql import functions as F

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "2023-01-01", 100),
    (1, "2023-01-02", 150),
    (1, "2023-01-05", 200),
    (2, "2023-02-10", 300),
    (2, "2023-02-11", 350),
    (3, "2023-03-01", 400)
]

df = spark.createDataFrame(data, ["customer_id", "order_date", "amount"]) \
          .withColumn("order_date", col("order_date").cast("date"))

df.show()
```

```python
join_conditions = ((F.col("df_a.customer_id") == F.col("df_b.customer_id")) & (F.datediff(F.col("df_b.order_Date"), F.col("df_a.order_date")) == 1))


df.alias("df_a").join(df.alias("df_b"), on=join_conditions).select(F.col("df_a.customer_id"), F.col("df_a.order_date"), F.col("df_b.order_date").alias("next_day_order_date")).show()
```
