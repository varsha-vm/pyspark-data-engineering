```python
#Time:O(N)
#Space:O(N)

from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql import Window

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "Alice", "IT", 100000),
    (2, "Bob", "IT", 90000),
    (3, "Charlie", "HR", 70000),
    (4, "David", "HR", 60000),
    (5, "Eve", "IT", 110000)
]

df = spark.createDataFrame(
    data,
    ["emp_id", "name", "department", "salary"]
)

df.show()
```

```python
window = Window.partitionBy("department")

df = df.withColumn("avg_salary", F.avg("salary").over(window))

df.select(F.col("emp_id"), F.col("department"), F.col("salary")).filter(F.col("salary") > F.col("avg_salary")).show()
```
